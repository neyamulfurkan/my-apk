```python
#for diabetes datasets

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
path2=r"C:\important\machine learning files\CSV files\mulitple disease prediction\diabetes.csv"
data2=pd.read_csv(path2)
```

```python
X=data2.drop(columns=['Outcome'], axis=1)
Y=data2['Outcome']
```

```python
X_train, X_test, Y_train, Y_test= train_test_split(
    X,Y, test_size=0.1,stratify=Y, random_state=2
)
```

```python
scaler=StandardScaler()
```

```python
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
model2=SVC(kernel='linear')
```

```python
model2.fit(X_train_scaled, Y_train)
```

```python
#prediction on test data
prediction= model2.predict(X_test_scaled)
```

```python
print(f'accuracy: {accuracy_score(Y_test, prediction)*100: .2f}%')
```

```python
#making a predictive system
input_data=np.asarray((5,166,72,19,175,25.8,0.587,51)).reshape(1,-1)
```
